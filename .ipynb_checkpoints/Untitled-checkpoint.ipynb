```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500.plot.line(y="Close",use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100, min_samples_split=100, random_state=1)
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
predictors = ["Close","Volume","Open","High","Low"]
```

```python
model.fit(train[predictors] , train["Target"])
```

```python
from sklearn.metrics import precision_score
preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds,index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    
    # Convert preds (numpy array) to a pandas Series
    preds_series = pd.Series(preds, index=test.index, name="Predictions")
    
    # Combine actual and predicted values
    combined = pd.concat([test["Target"], preds_series], axis=1)
    
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
```

```python
new_predictors = []
for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_columns = f"Close_Ratio_{horizon}"
    sp500[ratio_columns] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_columns, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds,index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
    
```

```python
predictions = backtest(sp500,model,new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
